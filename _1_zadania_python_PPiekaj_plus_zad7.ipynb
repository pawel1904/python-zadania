Napisz funkcję jeden_zero(rozmiar),
 która w minimalnej liczbie kroków tworzy (i zwraca) kwadratową tablicę
NumPy zawierającą na przemian jedynki i zera - w postaci jak na grafice -
 o dowolnym  rozmiarze (zadanym przez argument wywołania),
Powodzenia!


```python
import numpy as np
```

1 sposób

```python
def jeden_zero(dim):
    arr = np.ones((dim,dim))
    for col in range(arr.shape[1]):
        if col % 2 != 0:
            arr[:,col]=0
    return arr.astype("uint8")
```

```python
jeden_zero(7)
```

2 sposób

```python
def jeden_zero2(dim):
    return (np.zeros((dim,dim)) + np.array([1 if x % 2 == 0 else 0 for x in range(dim)])).astype("uint8")
```

```python
jeden_zero2(8)
```

3 sposób

```python
def jeden_zero3(dim):
    return np.full((dim,dim),np.array(list([1,0]*dim)[:dim]))
```

```python
jeden_zero3(5)
```

4 sposób

```python
def jeden_zero4(dim):
    temp = np.zeros((dim,dim))
    temp[:,::2] = 1
    return temp.astype("uint8")
```

```python
jeden_zero4(8)
```

```python
import pandas as pd
np.random.seed(123)
df = pd.DataFrame(np.random.randint(1,100, 100).reshape(10, -1))
df
```

1 sposób

```python
def zero_diag(df):
    for x in df.index.values:
        df.iloc[x,x] = df.iloc[x,-1*x-1] = 0
```

```python
zero_diag(df)
df
```

2 sposób

```python
def zero_diag2(df):
    np.fill_diagonal(df.values, 0)
    
    np.fill_diagonal(df.iloc[:,::-1].values, 0)
    #lub
    np.fill_diagonal(np.fliplr(df.values), 0)
```

```python
zero_diag2(df)
df
```

UWAGA - zadanie typowo algorytmiczne :slightly_smiling_face:
Wyobraźmy sobie tablicę o rozmiarach M na N utworzoną z wykorzystaniem numpy - wygenerować taką tablicę o wymiarach MxN.
W dowolnym, losowym jej miejscu "stawiamy" figurę skoczka szachowego.
Proszę napisać skrypt realizujący ruch owego "skoczka" po naszej
tablicy. Nasza figura może odwiedzić  dane pole tablicy tylko raz.
Kończymy działanie skryptu, kiedy już nie będzie możliwe wskoczenie na
kolejne pole. Pozycja wyjściowa skoczka jest dowolna.
Nie optymalizujemy drogi skoczka, nie musi przejść wszystkich pól.
Użyć metod do generowania wartości pseudolosowych. No i pamiętamy, że skoczek porusza się tak -> https://pl.wikipedia.org/wiki/Skoczek_(szachy)
Powodzenia!

```python
MOVES = [
    (2,1),
    (2,-1),
    (-2,1),
    (-2,-1),
    (1,2),
    (1,-2),
    (-1,2),
    (-1,-2)
]
```

```python
class chess_jumper:
    def __init__(self, m=8, n=8):
        self.m = m
        self.n = n
        self.chessboard = np.zeros((m,n))
        self.curr_row = np.random.randint(m)
        self.curr_col = np.random.randint(n)
        self.chessboard[self.curr_row, self.curr_col] = 8
    
    def avail_positions(self):
        self.positions = np.array(MOVES) + np.array([self.curr_row, self.curr_col])
        self.positions = self.positions[(self.positions[:,0] < self.m) & (self.positions[:,0] >=0) ]
        self.positions = self.positions[(self.positions[:,1] < self.n) & (self.positions[:,1] >=0) ]
        self.positions = np.array([x for x in self.positions if self.chessboard[x[0],x[1]] == 0 ])
    
    def move(self):
        self.chessboard[self.curr_row, self.curr_col] = 1
        temp = np.random.randint(self.positions.shape[0])
        self.curr_row = self.positions[temp,0]
        self.curr_col = self.positions[temp,1]
        self.chessboard[self.curr_row, self.curr_col] = 8
        
        
```

```python
c = chess_jumper(4,3)
```

```python
# 0 - pole nieodwiedzone
# 1 - pole odwiedzone
# 8 - aktualna pozycja
while (True):
    print(c.chessboard)
    c.avail_positions()
    if c.positions.shape[0] == 0:
        print("No more available moves")
        break
    c.move()
    input("Press Enter to continue...")
```

Uwaga - nie zawsze będzie możliwe pobranie wartości w gramach. Konwersję/przekształcenie należy wykonać w 2 odmianach:
• przy użyciu metody split()
• przy użyciu metody find()

```python
servings = pd.read_csv("https://www.dropbox.com/s/yruwnkgrqyegw78/Mcdonalds.csv?dl=1",  index_col=["Item"], usecols=["Item","Serving Size"])
```

```python
def convert_weight(x):
    temp = x.split("(")
    if len(temp) < 2:
        return np.nan
    temp = temp[1].split(")")
    if not(temp[0].endswith("g")):
        return np.nan
    return temp[0]
```

```python
def convert_weight2(x):
    l = x.find("(")
    if l == -1:
        return np.nan
    r = x.find(")", l)
    if r == -1:
        return np.nan

    if not(x[r-1].endswith("g")):
        return np.nan
    return x[l+1:r]
```

```python
%timeit -r 10 -n 100 [convert_weight(x) for x in servings["Serving Size"]]
```

```python
%timeit -r 10 -n 100 [convert_weight2(x) for x in servings["Serving Size"]]
```

```python
servings["Serving Size"] = servings["Serving Size"].apply(convert_weight) 
```
